```python
# importing libraries:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# displaying all rows and columns:
pd.set_option("display.max_columns",None)
pd.set_option("display.max_rows",None)

#for ignoring warnings:
import warnings
warnings.filterwarnings("ignore")
```

```python
# importing ipl dataset:
df=pd.read_csv(r"C:\Users\Vishal\Web Deployment\IPL Score Prediction\data\ipl.csv",header=0)
print(df.shape)
df.head(10)
```

```python
# checking all the details of the dataset:
df.info()
```

```python
# checking central tendency and quartile value using describe():
df.describe(include="all")
```

```python
# checking NA values using heatmap:
plt.figure(figsize=(15,5))
sns.heatmap(df.isnull(),yticklabels=False,cbar=False)
```

```python
# checking null value:
df.isnull().sum()
```

```python
# So as there is no missing value we can delete some columns which is not required eg.mid, batsman, bowler, striker and 
# non-stricker because their are many player in Indian Premeier League so we canoot know each and every player
colname=["mid","batsman","bowler","striker","non-striker"]
for i in colname:
    df.drop([i],axis=1,inplace=True)
    
df.columns
```

```python
# changing date column 'object' to 'datatime' datatype:
df["date"]=pd.to_datetime(df["date"])

print(df.dtypes)
```

```python
# So now lets check about all columns those who has data type as 'object' how many unique columns are available in each 
# columns
for i in df.columns:
    if df[i].dtypes==object:
        print(i,":")
        print(df[i].unique())
        print()
```

```python
# keeping only those teams who played all the seasons of ipl from 2008 to 2017:
consistent_teams=["Kolkata Knight Riders","Chennai Super Kings","Rajasthan Royals","Mumbai Indians","Kings XI Punjab",
                 "Royal Challengers Bangalore","Delhi Daredevils","Sunrisers Hyderabad"]

# to print "consistent_teams" in ascending order:
# print(pd.Series(consistent_teams).sort_values(ascending=True))

print(consistent_teams)
```

```python
df=df[(df["bat_team"].isin(consistent_teams))&(df["bowl_team"].isin(consistent_teams))]

print("bat_team:",df["bat_team"].sort_values(ascending=True).unique())
print()
print("bowl_team:",df["bowl_team"].sort_values(ascending=True).unique())
# sorting unique value of "bat_teams" and "bowl_teams" because it is easy to compare.
```

```python
#Removing 1st five overs from every match:
df=df[df['overs']>=5.0]
print(df.shape)
df.head()
```
